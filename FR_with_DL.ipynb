```python
# import libraries
from mtcnn.mtcnn import MTCNN
from matplotlib.patches import Rectangle, Circle
import matplotlib.pyplot as plt
import os
```

```python
# draw the detected faces on the squares
def draw_image_with_boxes(filename, result_list):
    #load the image
    data = plt.imread(os.path.join("test_images", filename))
    #plot the loaded image
    plt.imshow(data)
    #get the context for drawing boxes
    ax = plt.gca()
    #plot each box
    for square in result_list:
        #get coordinates
        x,y,width,height = square["box"]
        #create the shape
        rect = Rectangle((x,y),width, height, fill = False, color="red")
        #draw the box
        ax.add_patch(rect)
        #draw the landmark dots
        for key,value in square["keypoints"].items():
            #create and draw dot
            dot = Circle(value, radius=2, color="red")
            ax.add_patch(dot)
    #show final plot
    plt.show()
```

```python
#load image from file
filename = "test1.jpg"
pixels = plt.imread(os.path.join("test_images",filename))
```

```python
# Create model constructor usint default weights
detector = MTCNN()
#detect the faces in the image
faces = detector.detect_faces(pixels)
# display faces on the original image
draw_image_with_boxes(filename, faces)
```
