```python
import cv2
from matplotlib import pyplot as plt
```

```python
#check opencv version
print(cv2.__version__)
```

Downloade the haarcascade frontalface detection file from OpenCV GitHub -> 
https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml and save it in the working directory as  ‘haarcascade_frontalface_default.xml‘.

```python
# Load the test photographs
pixels = cv2.imread("test_images/test1.jpg")
```

```python
plt.imshow(pixels)
```

```python
# Load the pretrained model form OpenCV
classifier = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
# perform the face detection
bboxes = classifier.detectMultiScale(pixels)

#print the bounding box for each detected face
for box in bboxes:
    print(box)
    # extract and calculate the boxes parameters
    x,y,width, height = box
    x2,y2 = x + width, y + height
    
    #draw a rectangle over the pixels
    cv2.rectangle(pixels,(x,y), (x2,y2), (0,0,255),5)
    

```

```python
# show the image with the boxes
plt.imshow(pixels)
```
